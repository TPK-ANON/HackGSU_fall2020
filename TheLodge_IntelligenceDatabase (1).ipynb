```python
# Step 1. Dependencies

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
#load the data to lodge_data
lodge_data = pd.read_csv("Lodge_data.csv")
lodge_data
```

```python
#lodge_data.head(13)
lodge_data.tail(13)
```

```python
#split dataset in features and target variables
feature_cols = ['Company', 'Phobias'] 
X = lodge_data[feature_cols] # Features
Y = lodge_data.Phobias # Target variable
```

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import LeaveOneOut
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# import the class and create classifier/model
from sklearn.neural_network import MLPClassifier
for j in range(10000000): 
      cv = LeaveOneOut()
model = MLPClassifier(hidden_layer_sizes=(100, 50, 101),activation='relu',random_state=101)
```

```python
# split X and Y into training and testing sets
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.75,random_state=0)
```

```python
#train that model
   
for j in range(10): 
      cv = LeaveOneOut()
      model.fit(X_test,Y_test) 
```

```python
#predict
Y_predict = model.predict(X_test)
```

```python
# import confusion_matrix and classification_report classes
from sklearn.metrics import classification_report, confusion_matrix  
```

```python
#computer performance measures
print(confusion_matrix(Y_test, Y_predict))  
print(classification_report(Y_test, Y_predict))  
```

```python
from sklearn import metrics
#Calculate performance measures:
print("Accuracy:",metrics.accuracy_score(Y_test, Y_predict))
```
